```python
# Header:
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2
from decimal import Decimal

def format_e(n):
    a = '%e' % Decimal(n)
    return a.split('e')[0].rstrip('0').rstrip('.')+'e'+a.split('e')[1]
```

```python
d = np.array([9,
              8,
              7.144,
              6,
              5,
              4,
              3,
              2,
              1.5]) *1e-3
Fehler_d = d*0.01

t = np.array([2.87,3.01,3.12,3.14,3.12,
              3.86,4.48,3.81,3.96,3.95,
              18.00,12.81,13.15,13.23,12.89,
              5.87,5.84,5.80,5.86,6.01,
              7.78,7.78,7.87,7.92,8.07,
              11.58,11.12,6.09,5.75,5.87,
              40.15,40.46,40.40,40.42,40.65,
              19.20,20.07,38.32,40.50,39.68,
              91.84,95.61,87.18,99.28])
Fehler_t = np.full(t.size,0.4)       

h = np.array([100,100,100,100,100,
              100,100,100,100,100,
              400,300,300,300,300,
              100,100,100,100,100,
              100,100,100,100,100,
              100,100,50,50,50,
              200,200,200,200,200,
              50,50,100,100,100,
              150,150,150,150]) *1e-3
Fehler_h = np.full(h.size,3) *1e-3

rho = np.array([1.3625,
                1.3575,
                1.3775,
                1.3775,
                1.3775,
                1.3775,
                1.3775,
                1.3775,
                1.3925]) *1e3
Fehler_rho = np.full(rho.size,0.0025) *1e3

D = 75 *1e-3
Fehler_D = 0.5*1e-3

g = 9.80984
Fehler_g = 0.00002

rho_PEG1 = 1.1487 *1e3
Fehler_rho_PEG1 = 0.0054 *1e3

v_mean = np.zeros(9)
v_mean[0] = np.mean((h/t)[0:5])
v_mean[1] = np.mean((h/t)[5:10])
v_mean[2] = np.mean((h/t)[10:15])
v_mean[3] = np.mean((h/t)[15:20])
v_mean[4] = np.mean((h/t)[20:25])
v_mean[5] = np.mean((h/t)[25:30])
v_mean[6] = np.mean((h/t)[30:35])
v_mean[7] = np.mean((h/t)[35:40])
v_mean[8] = np.mean((h/t)[40:44])
Fehler_v_mean = np.zeros(9)
Fehler_v_mean[0] = np.sqrt(np.mean(((h/t)**2*((Fehler_h/h)**2+(Fehler_t/t)**2))[0:5]/5))
Fehler_v_mean[1] = np.sqrt(np.mean(((h/t)**2*((Fehler_h/h)**2+(Fehler_t/t)**2))[5:10]/5))
Fehler_v_mean[2] = np.sqrt(np.mean(((h/t)**2*((Fehler_h/h)**2+(Fehler_t/t)**2))[10:15]/5))
Fehler_v_mean[3] = np.sqrt(np.mean(((h/t)**2*((Fehler_h/h)**2+(Fehler_t/t)**2))[15:20]/5))
Fehler_v_mean[4] = np.sqrt(np.mean(((h/t)**2*((Fehler_h/h)**2+(Fehler_t/t)**2))[20:25]/5))
Fehler_v_mean[5] = np.sqrt(np.mean(((h/t)**2*((Fehler_h/h)**2+(Fehler_t/t)**2))[25:30]/5))
Fehler_v_mean[6] = np.sqrt(np.mean(((h/t)**2*((Fehler_h/h)**2+(Fehler_t/t)**2))[30:35]/5))
Fehler_v_mean[7] = np.sqrt(np.mean(((h/t)**2*((Fehler_h/h)**2+(Fehler_t/t)**2))[35:40]/5))
Fehler_v_mean[8] = np.sqrt(np.mean(((h/t)**2*((Fehler_h/h)**2+(Fehler_t/t)**2))[40:44]/4))

k = (1+2.1*(d/D))
Fehler_k = 2.1*(d/D)*np.sqrt((Fehler_d/d)**2+(Fehler_D/D)**2)

v_lam = k*v_mean
Fehler_v_lam = v_lam*np.sqrt((Fehler_v_mean/v_mean)**2+(Fehler_k/k)**2)
```

```python
breakpoint = 5
```

```python

x1 = (d/2)**2
y1 = v_lam/(rho-rho_PEG1)
Fehler_x1 = x1*2*(Fehler_d/d)
Fehler_y1 = y1*np.sqrt((Fehler_v_lam/v_lam)**2+(Fehler_rho**2+Fehler_rho_PEG1**2)/((rho-rho_PEG1)**2))

y2 = v_mean/(rho-rho_PEG1)
Fehler_y2 = y2*np.sqrt((Fehler_v_mean/v_mean)**2+(Fehler_rho**2+Fehler_rho_PEG1**2)/((rho-rho_PEG1)**2))

#Fitfunktion
from scipy import odr

def fit_func(p, x):
    (s) = p 
    return s*x

model = odr.Model(fit_func)

x = x1[5:x1.size]
y = y1[5:x1.size]
delta_x = Fehler_x1[5:x1.size]
delta_y = Fehler_y1[5:x1.size]

#Startparameter
para0 = [1]

data = odr.RealData(x, y, sx=delta_x, sy=delta_y)
odr = odr.ODR(data, model, beta0=para0 )
out = odr.run()

#1-Sigma
popt = out.beta
perr = out.sd_beta

#Sigma-Umgebung
nstd = 1 # um n-Sigma-Umgebung im Diagramm zu zeichnen

popt_top = popt+nstd*perr
popt_bot = popt-nstd*perr

#Plot-Umgebung
x_fit0 = np.linspace(min(x)-(max(x)-min(x))/10, max(x)+(max(x)-min(x))/10, 1000)
x_fit = np.linspace(min(x1)-(max(x1)-min(x1))/10, max(x1)+(max(x1)-min(x1))/10, 1000)
fit = fit_func(popt, x_fit)
fit_top = fit_func(popt_top, x_fit)
fit_bot = fit_func(popt_bot, x_fit)

fig, ax = plt.subplots(1, figsize=[6.4 *2, 4.8 *2])
plt.ticklabel_format(axis='both', style='sci', scilimits=(0,3), useMathText=True)
plt.errorbar(x1, y1, yerr=Fehler_y1, xerr=Fehler_x1, lw=1, ecolor='k', fmt='none', capsize=2, label='$v_{lam}$ (mit Ladenburgscher Korrektur)')
plt.errorbar(x1, y2, yerr=Fehler_y2, xerr=Fehler_x1, lw=1, ecolor='r', fmt='none', capsize=2, label='$v_{mean}$ (ohne Korrektur)')
plt.plot(x_fit, fit, 'r', lw=1, label='Fit')
plt.plot(x_fit0, np.zeros(x_fit0.size), 'g:', lw=5, label='linearer Bereich, wo der Fit durchgeführt wurde')
ax.fill_between(x_fit, fit_top, fit_bot, color='C3', alpha=.25, label=str(nstd)+r'$\sigma$'+'-Umgebung')
plt.title('Messung nach Stokes')
plt.grid(True)
plt.xlabel('$r^2$ [$m^2$]')
plt.ylabel('$v/({\\rho}-{\\rho}_{PEG})$ [$m^{4}$ $s^{-1}$ $kg^{-1}$]')
plt.legend(loc='best')
plt.savefig('figures/212_Fig1.pdf', format='pdf', bbox_inches='tight')

#Chi-Quadrat orthogonal
dof = x.size-popt.size
chisquare = out.sum_square
chisquare_red = chisquare/dof
prob = round(1-chi2.cdf(chisquare,dof),2)*100

#Auswertung
S = popt[0]
Fehler_S = perr[0]
eta1 = 2*g/(9*S)
Fehler_eta1 = eta1*np.sqrt((Fehler_g/g)**2+(Fehler_S/S)**2)

v_theo = (fit_func(popt, x1))*(rho-rho_PEG1)
Fehler_v_theo = v_theo*np.sqrt(((fit_func(popt+perr, x1)-fit_func(popt-perr, x1))/2/fit_func(popt, x1))**2+(Fehler_rho**2+Fehler_rho_PEG1**2)/((rho-rho_PEG1)**2))

Re = rho_PEG1*v_mean*d/eta1
Fehler_Re = Re*np.sqrt((Fehler_v_mean/v_mean)**2+(Fehler_rho_PEG1/rho_PEG1)**2+(Fehler_d/d)**2+(Fehler_eta1/eta1)**2)

#Ausgabe
print('Steigung [m^2 / (kg s)]: ')
print('S =', format_e(S), ' +- ', format_e(Fehler_S))
print('Chi-Quadrat =', chisquare)
print('Freiheitsgrade =', dof)
print('Chi-Quadrat reduziert =', chisquare_red)
print('Wahrscheinlichkeit ein größeres oder gleiches Chi-Quadrat zu erhalten =', prob, '%')
print('\n')                    
print('Viskosität: ')
print('eta_1 [Pa*s] =',  format_e(eta1), ' +- ',  format_e(Fehler_eta1))
print('\n')                    
print('Tabelle:')
i = 0
while i < 9:
    print('d [m] =',  format_e(d[i]), ' +- ',  format_e(Fehler_d[i]))
    print('v_mean [m / s] =',  format_e(v_mean[i]), ' +- ',  format_e(Fehler_v_mean[i]))
    print('v_theo [m / s] =',  format_e(v_theo[i]), ' +- ',  format_e(Fehler_v_theo[i]))
    print('Abweichung =', round((v_mean[i]-v_theo[i])/np.sqrt(Fehler_v_mean[i]**2+Fehler_v_theo[i]**2),2), 'sigma')
    print('Re =',  format_e(Re[i]), ' +- ',  format_e(Fehler_Re[i]))
    print('\n') 
    i = i+1
```

```python
k2 = 1/k
Fehler_k2 = k2*Fehler_k/k

x_fit1 = np.linspace(Re[0]*1.1, Re[breakpoint+1], 1000)
x_fit2 = np.linspace(Re[breakpoint-3], Re[Re.size-1]/1.1, 1000)

line1 = k2[0]+np.log(x_fit1/Re[0])*(k2[breakpoint-1]-k2[0])/np.log(Re[breakpoint-1]/Re[0])
line2 = k2[breakpoint]+np.log(x_fit2/Re[breakpoint])*(k2[k2.size-1]-k2[breakpoint])/np.log(Re[Re.size-1]/Re[breakpoint])

fig, ax = plt.subplots(1, figsize=[6.4 *2, 4.8 *2])
plt.ticklabel_format(axis='both', style='sci', scilimits=(0,3), useMathText=True)
plt.errorbar(Re, k2, yerr=Fehler_k2, xerr=Fehler_Re, lw=1, ecolor='k', fmt='none', capsize=2, label='Messdaten')
plt.plot(x_fit1, line1, 'r', lw=1, label='1. Gerade, turbulente Strömung')
plt.plot(x_fit2, line2, 'g', lw=1, label='2. Gerade, laminare Strömung')
plt.title('Bestimmung der kritischen Reynoldszahl ${Re}_{kr}$')
plt.xscale('log')
plt.grid(which='major')
plt.grid(which='minor', linestyle=':')
plt.xlabel('Re')
plt.ylabel('$v/v_{lam}$')
plt.legend(loc='best')
plt.savefig('figures/212_Fig2.pdf', format='pdf', bbox_inches='tight')
```

```python
V = np.array([5,10,15,20,25]) *1e-6
Fehler_V = np.full(V.size,0.2) *1e-6

t2 = np.array([142.42,286.19,426.84,576.70,721.65])
Fehler_t2 = np.full(t2.size,0.3) 

R = 1.5 *1e-3 /2
Fehler_R = 0.01 *1e-3 /2

L = 100 *1e-3
Fehler_L = 0.5 *1e-3 

h1 = 499 *1e-3
Fehler_h1 = 2 *1e-3

h2 = 495 *1e-3
Fehler_h2 = 2 *1e-3

q=np.array([0.2,0.4,0.6,0.8,1])

h_mean = h2 + (h1-h2)*q
Fehler_h_mean = np.sqrt((Fehler_h2*q)**2+(Fehler_h1*(1-q))**2) *1e-3

rho_PEG2 = 1.1476 *1e3
Fehler_rho_PEG2 = 0.0054 *1e3

eta2 = np.pi*(rho_PEG2*h_mean*g)*(R**4)*t2/(8*L*V)
Fehler_eta2 = eta2*np.sqrt((Fehler_g/g)**2+(Fehler_rho_PEG2/rho_PEG2)**2+(Fehler_h_mean/h_mean)**2
                           +4*(Fehler_R/R)**2+(Fehler_t2/t2)**2+(Fehler_L/L)**2+(Fehler_V/V)**2)

#Fitfunktion
from scipy import odr

def fit_func(p, x):
    (c) = p 
    return 0*x+c

model = odr.Model(fit_func)

x = np.array([1,2,3,4,5])
y = eta2
delta_x = np.array([1,1,1,1,1])
delta_y = Fehler_eta2

#Startparameter
para0 = [1]

data = odr.RealData(x, y, sx=delta_x, sy=delta_y)
odr = odr.ODR(data, model, beta0=para0 )
out = odr.run()

#1-Sigma
popt = out.beta
perr = out.sd_beta

#Sigma-Umgebung
nstd = 1 # um n-Sigma-Umgebung im Diagramm zu zeichnen

popt_top = popt+nstd*perr
popt_bot = popt-nstd*perr

#Plot-Umgebung
x_fit = np.linspace(min(x)-(max(x)-min(x))/10, max(x)+(max(x)-min(x))/10, 1000)
fit = fit_func(popt, x_fit)
fit_top = fit_func(popt_top, x_fit)
fit_bot = fit_func(popt_bot, x_fit)

fig, ax = plt.subplots(1, figsize=[6.4 *2, 4.8 *2])
plt.ticklabel_format(axis='both', style='sci', scilimits=(0,3), useMathText=True)
plt.errorbar(x, y, yerr=delta_y, lw=1, ecolor='k', fmt='none', capsize=10, label='Messdaten')
plt.plot(x_fit, fit, 'r', lw=1, label='Fit')
ax.fill_between(x_fit, fit_top, fit_bot, color='C3', alpha=.25, label=str(nstd)+r'$\sigma$'+'-Umgebung')
plt.title('Messung nach Hagen-Poiseuille')
plt.grid(True)
plt.xlabel('Messung Nr.')
plt.ylabel('${\eta}$ [$Pa$ $s$]')
plt.legend(loc='best')
plt.savefig('figures/212_Fig3.pdf', format='pdf', bbox_inches='tight')

#Chi-Quadrat orthogonal
dof = x.size-popt.size
chisquare = out.sum_square
chisquare_red = chisquare/dof
prob = round(1-chi2.cdf(chisquare,dof),2)*100

#Auswertung
eta2_mean = popt[0]
Fehler_eta2_mean = perr[0]

Re2 = 2*rho_PEG2*V/(t2*np.pi*R*eta2)
Fehler_Re2 = Re2*np.sqrt((Fehler_V/V)**2+(Fehler_rho_PEG2/rho_PEG2)**2+(Fehler_eta2/eta2)**2
                           +(Fehler_R/R)**2+(Fehler_t2/t2)**2)

Re2_mean = np.mean(Re2)
Fehler_Re2_mean = np.sqrt(np.mean(Fehler_Re2**2)/Re2.size)

#Ausgabe
print('Viskosität: ')
print('eta_2 [Pa*s] =',  format_e(eta2_mean), ' +- ',  format_e(Fehler_eta2_mean))
print('Chi-Quadrat =', chisquare)
print('Freiheitsgrade =', dof)
print('Chi-Quadrat reduziert =', chisquare_red)
print('Wahrscheinlichkeit ein größeres oder gleiches Chi-Quadrat zu erhalten =', prob, '%')
print('\n')  
print('Re =',  format_e(Re2_mean), ' +- ',  format_e(Fehler_Re2_mean))
```
